# Gaming Trends for Twitch

## Data analysis on the featured games on the video streaming site Twitch.

### Data has been extracted from SQL queries

* Import of modules

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

* Bar graph showing the viewing figures of the featured games

```python
plt.figure(figsize = (8,6))
games = ["LoL", "Dota 2", "CS:GO", "DayZ", "HOS", "Isaac", "Shows", "Hearth", "WoT", "Agar.io"]

viewers =  [1070, 472, 302, 239, 210, 171, 170, 90, 86, 71]
ax = plt.subplot()
plt.bar(range(len(games)), viewers)
plt.title('Viewing figures of featured games')
plt.legend(['Twitch'])
plt.xlabel('Featured games')
plt.ylabel('Viewing figures')
ax.set_xticks(range(len(games)))
ax.set_xticklabels(games)
plt.xticks(rotation=30)
plt.savefig('viewing_figures.png')
plt.show()
plt.clf()
```

* Pie chart showing the locations of League of Legends viewers

```python
plt.figure(figsize = (8,6))
labels = ["US", "DE", "CA", "N/A", "GB", "TR", "BR", "DK", "PL", "BE", "NL", "Others"]

countries = [447, 66, 64, 49, 45, 28, 25, 20, 19, 17, 17, 279]

colors = ['lightskyblue', 'gold', 'lightcoral', 'gainsboro', 'royalblue', 'lightpink', 'darkseagreen', 'sienna', 'khaki', 'gold', 'violet', 'yellowgreen']

plt.pie(countries, autopct='%d%%', pctdistance=1.1, colors=colors)
plt.axis('equal')
plt.title('Locations of League of Legends Viewers')
plt.legend(labels, loc='right')
plt.savefig('locations.png')
plt.show()
plt.clf()
```

* Viewing figures for January 1st 2015 in the UK. 15% error margins are shown.

```python
plt.figure(figsize = (8,6))
hour = range(24)

viewers_hour = [1, 4, 8, 11, 3, 1, 1, 0, 1, 1, 2, 12, 15, 17, 21, 17, 14, 8, 2, 5, 5, 2, 2, 0]

y_upper = []
y_lower = []
for data in viewers_hour:
  y_upper.append(data * 1.15)
  y_lower.append(data * 0.85)

plt.fill_between(hour, y_lower, y_upper, alpha=0.2)
plt.plot(hour, viewers_hour)
plt.title('Viewing figures for January 1st 2015 in the UK')
plt.xlabel('Time (hours)')
plt.ylabel('Viewing figures')
plt.savefig('uk_viewing_figures.png')
plt.show()
plt.clf()

```
